```python
import pandas as pd
import numpy as np

print("1. Create a simple DataFrame:")
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 35, 28, 32],
    'City': ['New York', 'London', 'Tokyo', 'Paris', 'Sydney'],
    'Salary': [50000, 60000, 70000, 55000, 65000],
    'Department': ['IT', 'Sales', 'IT', 'Marketing', 'Sales']
}
df = pd.DataFrame(data)
print(df)
print()

print("2. DataFrame info:")
print(df.info())
print()

print("3. Basic statistics:")
print(df.describe())
print()

print("4. First 2 rows:")
print(df.head(2))
print()

print("5. Last 2 rows:")
print(df.tail(2))
print()

print("6. Select Name column:")
print(df['Name'])
print()

print("7. Select multiple columns:")
print(df[['Name', 'Age', 'Salary']])
print()

print("8. Select row 1:")
print(df.iloc[1])
print()

print("9. Filter Age > 30:")
print(df[df['Age'] > 30])
print()

print("10. Filter IT department:")
print(df[df['Department'] == 'IT'])
print()

print("11. Sort by Salary (descending):")
print(df.sort_values('Salary', ascending=False))
print()

print("12. Average salary by department:")
print(df.groupby('Department')['Salary'].mean())
print()

print("13. Add Annual Salary column:")
df['Annual_Salary'] = df['Salary'] * 12
print(df)
print()

print("14. Create DataFrame with missing values:")
df_missing = df.copy()
df_missing.loc[1, 'Salary'] = np.nan
df_missing.loc[3, 'Age'] = np.nan
print(df_missing)
print()

print("15. Check for missing values:")
print(df_missing.isnull().sum())
print()

print("18. Rename columns:")
df_renamed = df.rename(columns={'Annual_Salary': 'Yearly_Income'})
print(df_renamed.columns.tolist())
print()

print("19. Drop column:")
df_dropped = df.drop(columns=['Annual_Salary'])
print(df_dropped.columns.tolist())
```
