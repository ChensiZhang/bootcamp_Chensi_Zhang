```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('C:/Users/Zhangchensi/bootcamp_Chensi_Zhang/project/data/processed/processed_data.csv')
df = df.iloc[:,1:]
df
```

```python
df['ln_open'] = np.log1p(df['open']) #dealing with right-skewed
df['ln_close'] = np.log1p(df['close']) #dealing with right-skewed
df['ln_high'] = np.log1p(df['high']) #dealing with right-skewed
df['ln_low'] = np.log1p(df['low']) #dealing with right-skewed
```

```python
df['open/close'] = df['open'] / df['close'] #capturing relative behavior, not absolute values,creating a stronger and more interpretable feature,mitigating multicollinearity issues.
df['high/low'] = df['high'] / df['low'] #capturing relative behavior, not absolute values,creating a stronger and more interpretable feature,mitigating multicollinearity issues.
```
