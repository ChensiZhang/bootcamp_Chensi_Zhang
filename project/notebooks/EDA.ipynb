```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt, seaborn as sns
from scipy.stats import skew, kurtosis
```

```python
data = pd.read_csv('C:/Users/Zhangchensi/bootcamp_Chensi_Zhang/project/data/processed/processed_data.csv')
data = data.iloc[:,1:]
data
```

```python
data.info
```

```python
desc = data[['open','high','low','close','volume']].describe().T
desc['skew'] = [skew(data[c].dropna()) for c in desc.index]
desc['kurtosis'] = [kurtosis(data[c].dropna()) for c in desc.index]
desc
```

```python
sns.histplot(data['open'], kde=True)
plt.title('open Distribution')
plt.show()
```

```python
sns.histplot(data['close'], kde=True)
plt.title('close Distribution')
plt.show()
```

```python
sns.histplot(data['high'], kde=True)
plt.title('high Distribution')
plt.show()
```

```python
sns.histplot(data['low'], kde=True)
plt.title('low Distribution')
plt.show()
```

```python
sns.boxplot(x=data['volume'])
plt.title('low (Outliers)')
plt.show()
```

```python
sns.scatterplot(data=data, x='open', y='close', hue='year')
plt.title('open vs close')
plt.show()
```

```python
sns.scatterplot(data=data, x='high', y='low')
plt.title('high vs low')
plt.show()
```
