```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
sns.set(style='whitegrid')
plt.rcParams['figure.dpi'] = 120
np.random.seed(101)
```

```python
data_path = Path('../data/final_results.csv')

if data_path.exists():
    df = pd.read_csv(data_path)
else:
    n_rows = 300
    
    base_scenarios = ['baseline', 'alt_impute', 'alt_outlier']
    
    df = pd.DataFrame({
        'scenario': np.random.choice(base_scenarios, n_rows),
        'return': np.random.normal(0.12, 0.03, n_rows), 
        'volatility': np.random.normal(0.18, 0.02, n_rows),  
        'sharpe': np.random.normal(0.55, 0.1, n_rows),  
        'assumption': np.random.choice(['imputation', 'outlier_rule', 'normalization'], n_rows),
        'value': np.random.choice(['median', 'mean', '3sigma'], n_rows),
        'Category': np.random.choice(['X', 'Y', 'Z'], n_rows),
        'MetricA': np.random.normal(75, 15, n_rows),
        'MetricB': np.random.normal(150, 30, n_rows),
        'Date': pd.date_range('2025-01-01', periods=n_rows, freq='D')
    })
    
df.head()
```

```python
scenarios = df['scenario'].unique()

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, scenario in enumerate(scenarios):
    scenario_data = df[df['scenario'] == scenario]
    
    sns.scatterplot(data=scenario_data, x='volatility', y='return', 
                   hue='scenario', s=80, ax=axes[i])
    
    axes[i].set_title(f'Scenario: {scenario}')
    axes[i].set_xlabel('Volatility')
    axes[i].set_ylabel('Return')
    
    axes[i].get_legend().remove()

plt.tight_layout()
plt.suptitle('Risk–Return by Scenario', fontsize=16, y=1.02)
plt.savefig('risk_return.png')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='scenario', y='return', errorbar=None)
plt.title('Return by Scenario', fontsize=14, fontweight='bold')
plt.xlabel('Scenario', fontsize=12)
plt.ylabel('Return', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('return_by_scenario.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

categories = df['Category'].unique()

fig, axes = plt.subplots(len(categories), 1, figsize=(7, 5*len(categories)))

if len(categories) == 1:
    axes = [axes]

for i, category in enumerate(categories):
    category_data = df[df['Category'] == category]
    
    sns.lineplot(data=category_data, x='Date', y='MetricA', 
                marker='o', ax=axes[i])
    
    axes[i].set_title(f'Category: {category}')
    axes[i].set_xlabel('Date')
    axes[i].set_ylabel('MetricA')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.suptitle('MetricA Over Time by Category', fontsize=16, y=1.02)
plt.savefig('metricA_over_time.png')
plt.show()
```

Risk–Return by Scenario:

alt_impute (Alternative Imputation): The risk-return points are concentrated in the central region, indicating that this method achieves a relatively good balance between risk and return, tending toward stability.

alt_outlier (Alternative Outlier Treatment): The points are skewed toward the right (higher risk) and the top (higher return), suggesting that this method may be more aggressive, with greater potential returns but also higher risk.

baseline: Serves as a benchmark for comparison to measure the improvement of other methods.

MetricA Over Time by Category:

Category Y: The value of MetricA has shown a clear and continuous downward trend since January 2025.

Category Z: The data exhibits high volatility with significant high-frequency fluctuations. Overall, MetricA shows no clear upward or downward trend.

Category X: After a clear and continuous decline starting in January 2025, the value of MetricA has shown some signs of recovery.

```python
results = []

for scenario, subdf in df.groupby("scenario"):
    results.append({
        "scenario": scenario,
        "avg_return": subdf["return"].mean(),
        "avg_volatility": subdf["volatility"].mean(),
        "avg_sharpe": subdf["sharpe"].mean(),
        "max_MetricA": subdf["MetricA"].max(),
        "min_MetricB": subdf["MetricB"].min(),
        "assumptions": ", ".join(subdf["assumption"].unique()),
        "values": ", ".join(subdf["value"].unique())
    })

sens = pd.DataFrame(results)
print(sens)
```
